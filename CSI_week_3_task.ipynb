```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Covid Data.csv')
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['AGE'], kde=True)
plt.title('Distribution of AGE')
plt.xlabel('AGE')
plt.ylabel('SEX')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='ICU', y='AGE', data=df)
plt.title('Scatter plot of AGE vs ICU')
plt.xlabel('ICU')
plt.ylabel('AGE')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='MEDICAL_UNIT', y='AGE', data=df)
plt.title('Box plot of DATE_DIED vs AGE')
plt.xlabel('MEDICAL_UNIT')
plt.ylabel('AGE')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.pairplot(df[['AGE', 'PATIENT_TYPE', 'MEDICAL_UNIT', 'SEX']])
plt.show()
```
